```python
from pathlib import Path

raw_root = Path("~/shared/birdclef/raw/birdclef-2025").expanduser()
! ls {raw_root}
soundscape_root = raw_root / "train_soundscapes"
! ls {soundscape_root} | head
soundscapes = sorted(soundscape_root.glob("*.ogg"))
display(len(soundscapes))
```

```python
import bioacoustics_model_zoo as bmz

bmz.list_models()
```

```python
for name, model_cls in bmz.list_models().items():
    print(f"{name}: {model_cls}")
    model = model_cls()
    # now we want to get prediction and embedding shapes
    pred = model.predict(soundscapes[:1])
    display(pred.head())
    display(pred.shape)
    emb = model.embed(soundscapes[:1])
    display(emb.head())
    display(emb.shape)
```
